### c. Generalize it:


```python
import re
result = 0

nums_dict = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9
    }


with open('input_2.txt', 'r') as data:
    inputs = data.readlines()
print(inputs)        


def find_alphabetical_numbers(line, nums_dict):
    res = []
    for num in nums_dict.keys():
        matches = [(match.start(), match.group()) for match in re.finditer(num, line)]
        if matches != []:
            for match in matches:
                res.append((match[0], str(nums_dict[num])))
    return res


for inp in inputs:
    first_digit = []
    last_digit = []
    for idx, c in enumerate(inp):
        if c.isdigit():
            first_digit.append((idx, c))
            break
    for idx, c in enumerate(inp[::-1]):
        if c.isdigit():
            last_digit.append(((len(inp)- (idx+1)), c))
            break

    res = find_alphabetical_numbers(inp, nums_dict)
    res.sort()

    if res != []:
        first_digit.append(res[0])
        last_digit.append(res[-1])

    print(first_digit)
    first_digit.sort()
    last_digit.sort()

    first_digit = first_digit[0][1]
    last_digit = last_digit[-1][1]
    
    s = int(first_digit + last_digit)
    result += s
    print((inp, s))


print(result)
```
